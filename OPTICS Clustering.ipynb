```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
#dataset
```

```python
X=dataset.iloc[ : ,3:5]
#X
```

```python
 from sklearn.cluster import OPTICS
clustering = OPTICS(min_samples=3,xi=0.3,min_cluster_size=3).fit(X)
y_OPT=clustering.fit_predict(X)
y_OPT
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['cluster-group']=y_OPT
supervised
```

```python
supervised.to_csv('cluster.csv',index=False)
```
