```python
import pandas as pd
import numpy
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
dependent = numpy.ravel(dependent)
#y_test = numpy.ravel(y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000]}
#'gamma':['auto','scale']
grid=GridSearchCV(SVR(), param_grid, refit=True,verbose=3, n_jobs=1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_

print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
prediction
```

```python
import pickle
filename="finalized_RFGRIDsearchCV_model.sav"
#writing the model in a new file
pickle.dump(grid,open(filename,'wb'))
```
