```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
#dataset
```

```python
X=dataset.iloc[ : ,3:5]
#X
```

```python
 from sklearn.cluster import AffinityPropagation
clustering = AffinityPropagation(damping=0.8,random_state=40,affinity='euclidean',convergence_iter=3).fit(X)
y_aff=clustering.fit_predict(X)
y_aff
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['cluster-group']=y_aff
supervised
```

```python
supervised.to_csv('cluster.csv',index=False)
```
