```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
#dataset
```

```python
X=dataset.iloc[ : ,3:5]
#X
```

```python
 from sklearn.cluster import MeanShift
clustering = MeanShift(bandwidth=25,n_jobs=20).fit(X)
y_MS=clustering.fit_predict(X)
y_MS
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['cluster-group']=y_MS
supervised
```

```python
supervised.to_csv('cluster.csv',index=False)
```
