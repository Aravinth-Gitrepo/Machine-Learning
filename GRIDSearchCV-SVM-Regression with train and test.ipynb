```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

x_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
#param_grid={'kernel':['linear', 'poly', 'rbf', 'sigmoid', 'precomputed'], 'gamma':['scale', 'auto'], 'C':[1.0,10,100,1000,10000]}
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(), param_grid, refit=True,verbose=3, n_jobs=1)
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_

print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
y_predict=grid.predict(x_test)
y_predict
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_predict)
r_score
```

```python
import pickle
filename="finalized_SVMGRIDsearchCV_model.sav"
#writing the model in a new file
pickle.dump(grid,open(filename,'wb'))
```
