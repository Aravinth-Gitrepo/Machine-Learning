```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
#dataset
```

```python
X=dataset.iloc[ : ,3:5]
#X
```

```python
from sklearn.cluster import SpectralClustering
clustering = SpectralClustering(n_clusters=8,assign_labels='cluster_qr',random_state=40).fit(X)
y_spec=clustering.fit_predict(X)
y_spec
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['cluster-group']=y_spec
supervised
```

```python
supervised.to_csv('cluster.csv',index=False)
```
